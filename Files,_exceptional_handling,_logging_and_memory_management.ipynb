9. What is the difference between multithreading and multiprocessing?

   Ans:- Introduction to Multithreading and Multiprocessing in Python ...Multithreading and multiprocessing are both techniques to improve the performance of a system by enabling concurrent execution, but they differ in how they achieve this. Multithreading involves running multiple threads within a single process, sharing the same memory space, while multiprocessing involves running multiple independent processes, each with its own memory space.

   Multithreading:
  
   Single Process, Multiple Threads:
   A single program is divided into multiple smaller units of execution called threads, which run concurrently within the same process.
  
   Shared Memory:
   All threads within a process share the same memory space, making it easier to share data between them.
  
   Faster Context Switching:
   Switching between threads is generally faster and less resource-intensive than switching between processes.

   Global Interpreter Lock (GIL):
   In some languages like Python, the GIL can limit the true parallelism of multithreading, meaning only one thread may be executing Python bytecode at a time.

   Suitable for I/O-bound tasks:
   Multithreading is often preferred when tasks involve waiting for I/O operations (like network requests or disk access), as threads can release the CPU while waiting.

   Multiprocessing:

   Multiple Processes:
   Multiple independent processes are created, each with its own memory space.

   Separate Memory:
   Each process has its own memory space, which provides better isolation and fault tolerance.
  
   Slower Context Switching:
   Switching between processes is generally slower and more resource-intensive than switching between threads.

   True Parallelism:
   Multiprocessing can achieve true parallelism, especially on systems with multiple CPU cores, as each process can run on a separate core.   
  
   Suitable for CPU-bound tasks:
   Multiprocessing is often preferred for CPU-intensive tasks where multiple cores can be utilized to speed up computation.
  
   In essence:
   Multithreading is like having multiple cooks working on different parts of the same dish in the same kitchen (shared ingredients).
   Multiprocessing is like having multiple chefs working on different dishes in separate kitchens (separate ingredients and resources).

11.  What is memory management in Python?

  Ans:- Memory management in Python refers to the system that handles the allocation and deallocation of memory for Python objects and data structures. Unlike languages like C or C++ where manual memory management is often required, Python employs an automatic system that largely frees the programmer from this responsibility.

  The core components of Python's memory management are:

  Private Heap:
  All Python objects and data structures reside in a private heap, managed by the Python interpreter. This heap is distinct from the system's general memory heap.

  Reference Counting:
  This is the primary mechanism for managing object lifetimes. Each Python object maintains a reference count, which tracks the number of references pointing to that object.
  When a new reference to an object is created (e.g., assigning it to a variable), the reference count increases.
  When a reference is deleted or goes out of scope, the reference count decreases.
  When an object's reference count drops to zero, it signifies that no part of the program is currently using or referencing that object.

12. What are the basic steps involved in exception handling in Python?

     Ans:- The basic steps involved in exception handling in Python using the try, except, else, and finally blocks are as follows:

     try Block:
     This block encloses the code that is anticipated to potentially raise an exception.
     If an exception occurs within this block, the execution of the try block is immediately stopped, and control is transferred to the appropriate except block.

     except Block(s):
     These blocks follow the try block and are responsible for catching and handling specific types of exceptions that might occur within the try block.
     Multiple except blocks can be used to handle different types of exceptions separately, providing specific error handling logic for each.
     A general except block can be used to catch any exception not caught by more specific except blocks.

     else Block (Optional):
     This block executes only if no exception was raised within the try block.
     It is used for code that should run when the try block completes successfully without any errors.

     finally Block (Optional):
     This block always executes, regardless of whether an exception occurred in the try block or was handled by an except block.
     It is typically used for cleanup operations, such as closing files or releasing resources, ensuring these actions happen even if an error occurs.

```python
try:
    # Code that might raise an exception
    result = 10 / 0  # This will raise a ZeroDivisionError
except ZeroDivisionError:
    # Handle the specific ZeroDivisionError
    print("Error: Cannot divide by zero!")
except ValueError:
    # Handle a different type of error
    print("Error: Invalid value provided.")
except Exception as e:
    # Catch any other unexpected exceptions
    print(f"An unexpected error occurred: {e}")
else:
    # This code runs if no exception occurred in the try block
    print("Operation successful!")
finally:
    # This code always runs, regardless of exceptions
    print("Cleanup operations completed.")
```

17. What is the difference between os.fork() and multiprocessing in Python?

  Ans:- 1. os.fork():
   Low-Level System Call:
   os.fork() is a direct wrapper around the Unix/POSIX fork() system call. When called, it creates a new process (the child) that is an almost exact copy of the parent process. This includes inheriting memory, open file descriptors, and other system resources.

   Platform Specific:
   os.fork() is only available on POSIX-compliant systems (like Linux, macOS, Unix). It does not work on Windows.

   Copy-on-Write (COW):
   The memory is not immediately duplicated but shared using a copy-on-write mechanism. This means that physical memory pages are only copied when one of the processes modifies them.

   Potential Issues with Multithreading:
   Forking a multithreaded process can be problematic and lead to unexpected behavior, as the child process inherits the parent's threads in an undefined state.

   2. multiprocessing Module:
   High-Level Abstraction:
   The multiprocessing module in Python provides a more abstract and user-friendly way to create and manage processes. It handles the complexities of inter-process communication (IPC) and resource management.

   Cross-Platform:
   The multiprocessing module is designed to work across different operating systems, including Windows, by using different "start methods" internally (e.g., fork on POSIX, spawn on Windows and macOS).

   Managed Resources:
   It offers features like Process objects, Queue for IPC, Pool for managing a fixed number of worker processes, and Manager for creating shared objects.

   Start Methods:
   It allows specifying how new processes are created using set_start_method() or get_context(), including fork, spawn, and forkserver. The spawn method is generally safer and more robust, especially when dealing with shared resources or multithreaded parent processes.

22. What are the challenges associated with memory management in Python?

  Ans:- Memory Leaks:
     Despite automatic garbage collection, memory leaks can occur, especially with circular references where objects reference each other, preventing their reference counts from dropping to zero and thus preventing deallocation by the garbage collector. Unclosed resources like file handles or network connections can also lead to leaks.

     Performance Overhead:
     The automatic nature of garbage collection and reference counting can introduce performance overhead, as the interpreter dedicates resources to managing memory, potentially leading to slower execution compared to languages with manual memory management.
     
     Memory Bloat and Fragmentation:
     Inefficient code or poor data structure choices can lead to high memory usage over time (memory bloat). Additionally, repeated allocation and deallocation of varying-sized objects can lead to memory fragmentation, where free memory is scattered in small, unusable chunks, impacting performance.
     
     Lack of Fine-Grained Control:
     Python's automatic memory management abstracts away the low-level details, which can be a disadvantage for applications requiring precise control over memory allocation and deallocation for performance-critical scenarios.
     
     Impact of the Global Interpreter Lock (GIL):
     While not directly a memory management challenge, the GIL can indirectly affect memory-intensive, multi-threaded applications by limiting true parallelism, potentially leading to increased memory usage in some scenarios as threads compete for resources.
     
     Debugging Memory Issues:
     Identifying and resolving memory-related issues like leaks or excessive consumption can be challenging due to the automatic nature of Python's memory management. Tools and profiling are often necessary to pinpoint the root cause.

6. Write a Python program that logs an error message to a log file when a division by zero exception occurs.

```python
import logging

logging.basicConfig(filename='division_errors.log', level=logging.ERROR,
                    format='%(asctime)s - %(levelname)s - %(message)s')

def safe_divide(numerator, denominator):
    """
    Performs division and logs an error if a ZeroDivisionError occurs.
    """
    try:
        result = numerator / denominator
        return result
    except ZeroDivisionError:
        logging.error("Attempted to divide by zero!")
        return None

# Example usage:
print(safe_divide(10, 2))
print(safe_divide(10, 0))
```

8.  Write a program to handle a file opening error using exception handling.

```python
def open_and_read_file(filename):
    """
    Attempts to open and read a file, handling potential FileNotFoundError
    and other IOErrors.
    """
    file_object = None  # Initialize file_object to None
    try:
        # Attempt to open the file in read mode ('r')
        file_object = open(filename, 'r')
        content = file_object.read()
        print(f"File '{filename}' opened and read successfully. Content:")
        print(content)
    except FileNotFoundError:
        # Handle the specific case where the file does not exist
        print(f"Error: The file '{filename}' was not found.")
    except IOError as e:
        # Handle other input/output errors (e.g., permission issues)
        print(f"An I/O error occurred while accessing '{filename}': {e}")
    except Exception as e:
        # Catch any other unexpected exceptions
        print(f"An unexpected error occurred: {e}")
    finally:
        # This block always executes, ensuring the file is closed if opened
        if file_object:
            file_object.close()
            print(f"File '{filename}' has been closed.")

# Example usage:
print("--- Attempting to open an existing file ---")
open_and_read_file("existing_file.txt") # Create this file for testing

print("\n--- Attempting to open a non-existent file ---")
open_and_read_file("non_existent_file.txt")

print("\n--- Attempting to open a file with restricted permissions (might need adjustment based on OS) ---")
# On Linux/macOS, try a path like "/root/restricted_file.txt" or similar
# On Windows, try a system path or a file with explicit permission restrictions
open_and_read_file("/dev/null_restricted_permissions.txt") # Example for Linux/macOS, will likely raise PermissionError
```

11. Write a Python program that uses a try-except block to handle an error when attempting to access a
dictionary key that doesn't existF

```python
def safe_dictionary_access(data_dict, key):
    """
    Attempts to access a dictionary key and handles KeyError if the key is not found.

    Args:
        data_dict (dict): The dictionary to access.
        key: The key to look for in the dictionary.

    Returns:
        The value associated with the key if found, otherwise a message indicating
        the key was not found.
    """
    try:
        value = data_dict[key]
        return f"The value for '{key}' is: {value}"
    except KeyError:
        return f"Error: The key '{key}' was not found in the dictionary."

# Example usage
my_dict = {
    "name": "Alice",
    "age": 30,
    "city": "New York"
}

# Accessing an existing key
print(safe_dictionary_access(my_dict, "name"))

# Attempting to access a non-existent key
print(safe_dictionary_access(my_dict, "country"))

# Another example with a different non-existent key
print(safe_dictionary_access(my_dict, "occupation"))
```

15.  Write a Python program that prints the content of a file and handles the case when the file is empty.

```python
import os

def print_file_content(filepath):
    """
    Prints the content of a file.
    Handles FileNotFoundError if the file doesn't exist
    and prints a message if the file is empty.
    """
    try:
        if not os.path.exists(filepath):
            print(f"Error: File '{filepath}' not found.")
            return

        # Check if the file is empty using os.path.getsize()
        if os.path.getsize(filepath) == 0:
            print(f"The file '{filepath}' is empty.")
        else:
            with open(filepath, 'r') as file:
                content = file.read()
                print(f"Content of '{filepath}':\n{content}")

    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage:
# Create a dummy file for testing
with open("test_file.txt", "w") as f:
    f.write("This is a test file.\n")
    f.write("It contains multiple lines.")

with open("empty_file.txt", "w") as f:
    pass # Creates an empty file

print_file_content("test_file.txt")
print_file_content("empty_file.txt")
print_file_content("non_existent_file.txt")

# Clean up dummy files
os.remove("test_file.txt")
os.remove("empty_file.txt")
```

17. Write a Python program to create and write a list of numbers to a file, one number per line.

```python
def write_numbers_to_file(numbers, filename="output.txt"):
    """
    Writes a list of numbers to a file, with each number on a new line.

    Args:
        numbers (list): A list of numbers (integers or floats).
        filename (str): The name of the file to write to.
    """
    try:
        with open(filename, 'w') as file:
            for number in numbers:
                file.write(str(number) + '\n')
        print(f"Numbers successfully written to {filename}")
    except IOError as e:
        print(f"Error writing to file: {e}")

if __name__ == "__main__":
    # Example usage:
    my_numbers = [10, 25, 3.14, 42, 5.0, 600]
    write_numbers_to_file(my_numbers, "my_numbers.txt")

    another_list = list(range(1, 11)) # Creates a list from 1 to 10
    write_numbers_to_file(another_list, "sequential_numbers.txt")
```

18. How would you implement a basic logging setup that logs to a file with rotation after 1MB?

```python
import logging
from logging.handlers import RotatingFileHandler

def setup_rotating_logger(log_file_path, max_bytes=1024*1024, backup_count=5):
    """
    Sets up a logger that logs to a file with size-based rotation.

    Args:
        log_file_path (str): The path to the log file.
        max_bytes (int): The maximum size of the log file in bytes before rotation.
                         Default is 1MB (1024 * 1024 bytes).
        backup_count (int): The number of backup log files to keep.
    """
    logger = logging.getLogger(__name__)
    logger.setLevel(logging.INFO)  # Set the desired logging level

    # Create a RotatingFileHandler
    # maxBytes: rotate after this size (1MB)
    # backupCount: keep up to 5 old log files
    handler = RotatingFileHandler(log_file_path, maxBytes=max_bytes, backupCount=backup_count)

    # Define the log message format
    formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
    handler.setFormatter(formatter)

    # Add the handler to the logger
    logger.addHandler(handler)

    return logger

if __name__ == "__main__":
    # Example usage:
    logger = setup_rotating_logger("app.log")

    logger.info("This is an informational message.")
    logger.warning("This is a warning message.")
    logger.error("This is an error message.")

    # Simulate writing a lot of logs to trigger rotation
    for i in range(20000): # Adjust this number based on log message size to trigger rotation
        logger.info(f"Log message number {i}: This is a long message to fill up the log file quickly. " * 5)
```
